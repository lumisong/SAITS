```python
import os
import configparser
from collections import defaultdict
import pandas as pd

def read_ini_file(file_path):
    config = configparser.ConfigParser()
    config.read(file_path)

    file_dict = defaultdict(dict)
    for section in config.sections():
        for key in config[section]:
            file_dict[section][key] = config[section][key]

    return file_dict

def read_ini_files_in_folder(folder_path):
    files_dict = {}

    for root, dirs, files in os.walk(folder_path):
        for file in files:
            if file.endswith(".ini"):
                file_path = os.path.join(root, file)
                file_dict = read_ini_file(file_path)
                file_dict['filename'] = file  # adding filename as a key
                files_dict[file] = file_dict

    return files_dict

# testing the functions
folder_path = '../configs'  # set your folder path here
ini_files_dict = read_ini_files_in_folder(folder_path)


# Convert each dictionary to a DataFrame and store them in a list
df_list = [pd.json_normalize(ini_files_dict[key]) for key in ini_files_dict]

# Concatenate all DataFrames in the list
df = pd.concat(df_list, ignore_index=True)

# Display the DataFrame
df
```

```python
df.to_excel('config_info.xlsx', index=False)
```
